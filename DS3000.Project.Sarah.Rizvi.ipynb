```python
import requests
import pandas as pd

# Replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
## only using ticker = MSFT 
## only looking at data between jan-26-23
url = 'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=MSFT&time_from=20230126T0000&time_to=20230226T0000&apikey=AJJRVX46T5F9OBQG'
r = requests.get(url)
data = r.json()
feed_data = data['feed']
feed_df = pd.DataFrame(feed_data)
feed_df
```

```python
feed_df['ticker_sentiment'][0]
```

```python
# Create lists to store data from original dataframe 
ticker_symbols = []
ticker_sentiment_scores = []
ticker_sentiment_labels = []
time_published = []
overall_sentiment_scores = []
overall_sentiment_labels = []


# Iterate over each row in the original dataframe 
for index, row in feed_df.iterrows():
    time = row['time_published']
    overall_sentiment_score = row['overall_sentiment_score']
    overall_sentiment_label = row['overall_sentiment_label']
    ticker_sentiments = row['ticker_sentiment']
    
    # Iterate over ticker sentiments for each row
    for sentiment in ticker_sentiments:
        ticker_symbols.append(sentiment['ticker'])
        ticker_sentiment_scores.append(sentiment['ticker_sentiment_score'])
        ticker_sentiment_labels.append(sentiment['ticker_sentiment_label'])
        time_published.append(time)

# Add data into the new dataframe 
new_df = pd.DataFrame({
    'ticker_symbol': ticker_symbols,
    'ticker_sentiment_score': ticker_sentiment_scores,
    'ticker_sentiment_label': ticker_sentiment_labels,
    'time_published': time_published,
    'overall_sentiment_score': overall_sentiment_score,
    'overall_sentiment_label': overall_sentiment_label
})

# Display new DataFrame
new_df
```

```python
new_df_msft = new_df[new_df['ticker_symbol'] == 'MSFT']
new_df_msft
```

```python
# Calculate Sentiment Scores per ticker per day 

# convert time_published column to date time 
new_df_msft['time_published'] = pd.to_datetime(new_df_msft['time_published'], format='%Y%m%dT%H%M%S')
new_df_msft['date_published'] = new_df_msft['time_published'].dt.strftime('%m-%d-%Y') # create new column with only the date 

# get average sentiment scores per day by adding column 
avg_sentiment = new_df_msft.groupby(['ticker_symbol', 'date_published'])['overall_sentiment_score'].mean().reset_index()
new_df_msft = pd.merge(new_df_msft, avg_sentiment, on=['ticker_symbol', 'date_published'], suffixes=('', '_avg'))
new_df_msft
```

```python
from yahoo_fin.stock_info import get_data
msft_jan_feb= get_data("msft", start_date="01/26/2023", end_date="02/29/2024", index_as_date = True, interval="1d")
# reset index to create time column
msft_jan_feb = msft_jan_feb.reset_index()
# rename index column to time column 
msft_jan_feb = msft_jan_feb.rename(columns={"index": "time"})
msft_jan_feb 
```

```python
# join data sets by msft_jan_feb['time'] and new_df_msft['date_published']
msft_jan_feb['time'] = pd.to_datetime(msft_jan_feb['time'])
new_df_msft['date_published'] = pd.to_datetime(new_df_msft['date_published'], format='%m-%d-%Y')
merged_df = pd.merge(msft_jan_feb, new_df_msft, left_on='time', right_on='date_published', how='inner')

merged_df
```
